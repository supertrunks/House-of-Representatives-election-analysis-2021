```python
import pandas as pd
```

```python
data = pd.read_csv("./House-of-Representatives-election-analysis-2021/House-of-Representatives-election-analysis-2021-by-Candidates.csv")
data
```

```python
partyTable=pd.read_csv("./House-of-Representatives-election-analysis-2021/PartyTable.csv")
```

```python
partyTable
```

```python
data.Party = data.Party.map(partyTable.set_index('PartyName').Code)
data
```

```python
data
```

```python
result = pd.DataFrame({
  'result': ['Win', 'Lose'],
  'code': [1, 0]})
```

```python
data.Result = data.Result.map(result.set_index('result').code)
data
```

```python
data[data["Result"]==1]
```

```python
!pip install seaborn-analyzer
import pandas as pd
from seaborn_analyzer import CustomPairPlot
```

```python
data[data["Result"]==1]
```

```python
OBJECTIVE_VARIALBLE = 'Result'#目的変数
EXPLANATORY_VALIABLES = ['Party','Q1','Q2','Q3','Q4','Q5', 'Q6','Q7','Q8','Q9','Q10']#説明変数
use_cols = [OBJECTIVE_VARIALBLE] + EXPLANATORY_VALIABLES
gp = CustomPairPlot()
gp.pairanalyzer(data[use_cols], hue='Result')
```
